```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import matplotlib.dates as mdates
```

```python
from fredapi import Fred
fred = Fred(api_key='f0b1174f8adb143e512d23d713b002e8')
```

```python
#South: 

# Define the series IDs
series_ids = ['CRXRNSA', 'DAXRSA', 'TPXRSA', 'MIXRSA', 'ATXRSA']

# Initialize an empty DataFrame to store the data
combined_df = pd.DataFrame()

# Retrieve and combine data for each series
for series_id in series_ids:
    data = fred.get_series(series_id)
    df = pd.DataFrame({'Date': data.index, f'Value_{series_id}': data.values})
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)  # Set Date as the index
    combined_df = pd.concat([combined_df, df], axis=1)

# Filter the DataFrame to keep only the data for the last year
last_year_data = combined_df[combined_df.index >= combined_df.index.max() - pd.DateOffset(years=1)]

# Filter the DataFrame to keep only the data for the last five years
last_five_years_data = combined_df[combined_df.index >= combined_df.index.max() - pd.DateOffset(years=5)]

# Create the plot for one-year x-axis
plt.figure(figsize=(10, 6))
for series_id in series_ids:
    plt.plot(last_year_data.index, last_year_data[f'Value_{series_id}'], label=series_id)

plt.xlabel('Date')
plt.ylabel('Value')
plt.title(f'{", ".join(series_ids)} Over the Last Year')
plt.legend()
plt.grid(True)

# Set the x-axis ticks and labels to display only the months
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))

# Show the plot
plt.show()

# Create the plot for five-year x-axis
plt.figure(figsize=(10, 6))
for series_id in series_ids:
    plt.plot(last_five_years_data.index, last_five_years_data[f'Value_{series_id}'], label=series_id)

plt.xlabel('Date')
plt.ylabel('Value')
plt.title(f'{", ".join(series_ids)} Over the Last Five Years')
plt.legend()
plt.grid(True)

# Set the x-axis ticks and labels to display only the years
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator(base=1))
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))

# Show the plot
plt.show()
```
